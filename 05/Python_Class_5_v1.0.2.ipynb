# Number of Words and Repetition

```python
f = open("quick.txt")
text = f.read()
```

```python
text
```

```python
words = text.split()
```

```python
words
```

```python
len(words)
```

```python
words
unwanted_chars = ".,!"
wordfreq = {}
for raw_word in words:
    word = raw_word.strip(unwanted_chars)
    print(word)
    if word not in wordfreq:
        wordfreq[word] = 0 # 
    wordfreq[word] += 1
    print(wordfreq)
```

# Number of Alphabets

```python
f = open("quick.txt")
text = f.read()
```

```python
text1 = text.replace("\n","")
text2 = text1.replace(".","")
text3 = text2.replace(" ","")
text4 = text3.replace("!","")
text5 = text4.replace(",","")
text6 = text5.lower()
```

```python
text1
```

```python
text2
```

```python
text3
```

```python
text4
```

```python
text5
```

```python
text6
```

```python
character_frequency = {}
for letter in list(text6):
    if  letter not in character_frequency:
        character_frequency[letter] = 1
    else:
        character_frequency[letter] += 1
```

## Pandas

Think of Panda as a tool that picks functions and code already defined and labored in NumPy to save the user to time and space to handle big data. Very simply put Panda is the cake and NumPy are all the ingredients that need to put together to make that cake. Large scale you need a machine that cranks output instantly rather then hand making the product.
Panda is a library that has its root dating back to 2008 that begun its inception by a former employee of AQR Capital Management Wes McKinney, who wanted to accelerate quantitative analysis. AQR Capital was generous enough to allow the code to be open sourced.
Pandas library allows us to do data anlysis, manipulation and storing very easily in python.
